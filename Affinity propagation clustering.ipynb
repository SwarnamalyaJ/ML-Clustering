```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
```

```python
dataset
```

```python
X = dataset.iloc[:, 3:5].values
```

```python
X
```

```python
from sklearn.cluster import AffinityPropagation
aff = AffinityPropagation(preference=-10000, random_state=35) 
```

```python
y_aff = aff.fit_predict(X)
```

```python
y_aff
```

```python
supervised=dataset
```

```python
supervised
```

```python
supervised['Cluster_group']=y_aff
```

```python
supervised
```

```python
y_aff
```

```python
supervised.columns
```

```python
import seaborn as sns
# Plotting
facet = sns.lmplot(data=supervised, x=supervised.columns[3], y=supervised.columns[4], 
                   hue='Cluster_group', fit_reg=False, legend=True, legend_out=True)
```
