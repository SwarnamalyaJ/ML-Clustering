```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
```

```python
dataset
```

```python
X = dataset.iloc[:, 3:5].values
```

```python
X
```

```python
from sklearn.cluster import AgglomerativeClustering
agg_clustering = AgglomerativeClustering(n_clusters=5)
y_agg = agg_clustering.fit_predict(X)
```

```python
supervised_agg = dataset.copy()
supervised_agg['Cluster_group'] = y_agg
supervised_agg.to_csv("cluster_agglomerative.csv", index=False)
```

```python
# Visualize the clusters
plt.figure(figsize=(10, 6))
plt.scatter(X[y_agg == 0, 0], X[y_agg == 0, 1], s=50, c='blue', label='Cluster 1')
plt.scatter(X[y_agg == 1, 0], X[y_agg == 1, 1], s=50, c='green', label='Cluster 2')
plt.scatter(X[y_agg == 2, 0], X[y_agg == 2, 1], s=50, c='orange', label='Cluster 3')
plt.scatter(X[y_agg == 3, 0], X[y_agg == 3, 1], s=50, c='red', label='Cluster 4')
plt.scatter(X[y_agg == 4, 0], X[y_agg == 4, 1], s=50, c='purple', label='Cluster 5')

plt.title('Agglomerative Clustering')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.legend()
plt.show()
```

```python
facet = sns.lmplot(data=supervised_agg, x=supervised_agg.columns[3], y=supervised_agg.columns[4], hue='Cluster_group',
                   fit_reg=False, legend=True, legend_out=True)
```
