```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
```

```python
dataset
```

```python
X = dataset.iloc[:, 3:5].values
```

```python
X
```

```python
from sklearn.cluster import AffinityPropagation
from sklearn.preprocessing import StandardScaler
X_scaled = StandardScaler().fit_transform(X)
affinity_propagation = AffinityPropagation(preference=-50)  
y_affinity = affinity_propagation.fit_predict(X_scaled)
```

```python
supervised_affinity = dataset.copy()
supervised_affinity['Cluster_group'] = y_affinity
supervised_affinity.to_csv("cluster_affinity.csv", index=False)
```

```python
# Visualize the clusters
plt.figure(figsize=(10, 6))
plt.scatter(X_scaled[y_affinity == -1, 0], X_scaled[y_affinity == -1, 1], s=50, c='red', label='Noise')
plt.scatter(X_scaled[y_affinity != -1, 0], X_scaled[y_affinity != -1, 1], s=50, c=y_affinity[y_affinity != -1], cmap='viridis')
plt.title('Affinity Propagation Clustering')
plt.xlabel('Annual Income (scaled)')
plt.ylabel('Spending Score (scaled)')
plt.legend()
plt.show()
```

```python
#Using seaborn to plot with cluster labels
facet_affinity = sns.lmplot(data=supervised_affinity, x=supervised_affinity.columns[3], y=supervised_affinity.columns[4], hue='Cluster_group',
                             fit_reg=False, legend=True, legend_out=True)
```
