```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
```

```python
dataset
```

```python
X = dataset.iloc[:, 3:5].values
```

```python
X
```

```python
from sklearn.cluster import Birch
birch = Birch(n_clusters=5)  
y_birch = birch.fit_predict(X)
```

```python
supervised_birch = dataset.copy()
supervised_birch['Cluster_group'] = y_birch
supervised_birch.to_csv("cluster_birch.csv", index=False)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X[y_birch == 0, 0], X[y_birch == 0, 1], s=50, c='blue', label='Cluster 1')
plt.scatter(X[y_birch == 1, 0], X[y_birch == 1, 1], s=50, c='green', label='Cluster 2')
plt.scatter(X[y_birch == 2, 0], X[y_birch == 2, 1], s=50, c='orange', label='Cluster 3')
plt.scatter(X[y_birch == 3, 0], X[y_birch == 3, 1], s=50, c='red', label='Cluster 4')
plt.scatter(X[y_birch == 4, 0], X[y_birch == 4, 1], s=50, c='purple', label='Cluster 5')

plt.title('Birch Clustering')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.legend()
plt.show()
```

```python
facet_birch = sns.lmplot(data=supervised_birch, x=supervised_birch.columns[3], y=supervised_birch.columns[4], hue='Cluster_group',
                         fit_reg=False, legend=True, legend_out=True)
```
